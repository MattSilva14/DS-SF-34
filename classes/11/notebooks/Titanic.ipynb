```python
df = df.dropna()
```

```python
X = df[ ['Age', 'Pclass', 'Fare', 'poorguy','richguy', 'PassengerId']]
c = df.Survived

model = linear_model.LogisticRegression(penalty = 'l1').\
    fit(X, c)
```

```python
model.score(X, c)
```

```python
c_hat = model.predict(X)

pd.crosstab(c_hat,
    c,
    rownames = ['Hypothesized Class'],
    colnames = ['True Class'])
```
